## Import matplotlib / Importando matplotlib

One of the easiest ways to make use of this library is through the `pyplot` module, into which we usually import it under the alias` plt`. 

We will also be using numpy to initially create a dataset for visualization purposes.

---

Una de las formas más sencillas de hacer uso de esta librería es a través del módulo `pyplot`, al cual lo solemos importar bajo el alias `plt`.

Vamos a usar numpy para crear un set de datos que vamos a visualizar. 

```python
import matplotlib.pyplot as plt
import numpy as np 
```

## Plot functions / Graficando de funciones 

Many times we want to graphically visualize a mathematical function that depends on a certain variable.
For example, let's graph the sine function between 0 and 2 $\pi$.

---

Muchas veces deseamos visualizar gráficamente una función matemática que depende de cierta variable.
Por ejemplo, grafiquemos la función seno entre 0 y 2$\pi$.

```python
x = np.linspace(0, 2 * np.pi, 100)
print(x)
```

```python
y = np.sin(x)
```

```python
plt.plot(x, y)
plt.show()
```

```python
plt.plot(x, y)

plt.xlabel('Angle')
plt.ylabel('Sine')

plt.show()
```

## Multiple curves in the same plot / Múltiples curvas en un mismo gráfico

How can we include several curves in the same graph? / ¿Cómo podemos incluir varias curvas en un mismo gráfico?

```python
plt.plot(x, np.sin(x), linestyle="-", label="sin(x)")
plt.plot(x, np.sin(2 * x), linestyle="--", label="sin(2x)")
plt.plot(x, np.sin(3 * x), linestyle="-.", label="sin(3x)")

plt.legend()

plt.xlabel('Angle')

plt.show()
```

## Good practices / Buenas prácticas

When we want to plot and publish data, we always have to make sure that our figures have:

- Annotations of the coordinate axes
- Axis units
- Legend (if we have several plots in the same figure)

In addition, we can add information and some details such as:

- A title
- Decorations, like a grid

---

A la hora de visualizar datos y publicarlos, siempre tenemos que asegurarnos de que nuestras figuras cuenten con:

- Anotaciones de los ejes coordenados
- Unidades de los ejes
- Leyenda (si tenemos varios plots en la misma figura)

Además, podemos añadir información y algunos detalles como:

- Un título
- Decoraciones como un grid

```python
plt.plot(x, np.sin(x), linestyle="-", label="sin(x)")
plt.plot(x, np.sin(2 * x), linestyle="--", label="sin(2x)")
plt.plot(x, np.sin(3 * x), linestyle="-.", label="sin(3x)")

plt.legend()

plt.xlabel("Angle")
plt.ylabel("y")

plt.grid()

plt.show()
```

## Plot data / Graficando datos

Now we will plot the mean temperature data for entire planet, from 1850 to the present.
It is available by [Berkeley Earth](http://berkeleyearth.org/).
The data files can be downloaded from their website, but in the repository we already have a simplified version of one of them in `data/temperature-berkeley.dat`.

We are going to import the `Path` class from` pathlib` which helps us to create paths to files and folders on our disk.
Let's define the path to the temperatures:
* Using `Path` to this, we ensure that this code can work on various operating systems.
  No matter what type of directory separator they use (`/` or `\`).
* Remember that the symbol `..` refers to the parent directory.

---

Vamos a graficar los datos de temperaturas medias anual de toda el planeta, desde 1850 hasta la actualidad. 
Estos datos estan a disponibilidad por [Berkeley Earth](http://berkeleyearth.org/).
Los archivos de datos pueden ser descargados desde su página web, pero en el repositorio ya contamos con una versión simplificada de uno de ellos en `data/temperature-berkeley.dat`.

Vamos a importar la clase `Path` de `pathlib` que nos sirve para crear rutas a archivos y carpetas de nuestro disco.
Definamos la ruta al archivo de temperaturas:
* Al utilizar `Path` para esto, nos aseguramos que este código pueda funcionar en diversos sistemas operativos, sin importar qué tipo de separador de directorios utilizan (`/` o `\`) entre otras diferencias.
* Recordemos que el símbolo `..` hace referencia al directorio padre.


```python
from pathlib import Path
```

```python
berkeley_data_file = Path(".")/ "data" / "temperature-berkeley.dat"
berkeley_data_file
```

Let's use the `numpy.loadtxt` function to read the file / Usemos la función `numpy.loadtxt` para leer el archivo.

```python
datos = np.loadtxt(berkeley_data_file)
datos
```

We can extract the columns corresponding years and temperatures from the `data` array / Podemos extraer las columnas correspondiente años y temperaturas del array `datos`

```python
years, temperaturas = datos[:, 0], datos[:, 1]
```

```python
plt.plot(years, temperaturas, "o-")
plt.show()
```

We can save the plot as a `png` image or `PDF` (vectorial image) / Podemos guardar el gráfico como una imagen `png` o` PDF` (imagen vectorial)

```python
plt.plot(years, temperaturas, "o")
plt.xlabel("Años")
plt.ylabel("Temperatura [C]")
plt.title("Temperatura media anual de la Tierra")
plt.grid()

plt.savefig("temperatura2.png", dpi=300)
plt.savefig("eemperatura.pdf")

plt.show()
```

## Other type of plot / Otros tipos de gráficos

Matplotlib allows us to make multiple types of plots / Matplotlib nos permite realizar múltiples tipos de gráficos.

### Scatter

In case our data is scattered / En caso de que nuestros datos se encuentren dispersos.

```python
plt.scatter(years, temperaturas)
plt.show()
```

### Histrogram / Histogramas

```python
plt.hist(temperaturas)
plt.xlabel("Temperaturas [C]")
plt.ylabel("Frecuencia")
plt.show()
```

### pcolormesh

```python
# Data to plot.
x_aux = np.linspace(-np.pi, np.pi, 100)
# Create a grid / Creo una grilla
x, y = np.meshgrid(x_aux, x_aux)

# Calculate the values / Calculo los valores
z = np.sinc(x**2 + y**2)
```

```python
# Plot
plt.pcolormesh(x, y , z, shading="auto", cmap='RdBu', vmin=-1, vmax=1)
plt.colorbar()
```
